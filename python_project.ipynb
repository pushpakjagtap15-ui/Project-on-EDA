```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import Image

# display(Image("/content/GTA6.jpg", width=900))
print("""\nProblem statement: Rockstar Games is preparing for the global launch of GTA 6. The goal is to analyze
player engagement, spending habits, and gaming trends to optimize monetization,
server infrastructure, and marketing strategies..""")
print("* Note: Visual assets (e.g., GTA VI poster) are used only during presentation for illustrative purposes.\n")

# Dataset File : online_gaming_behavior_dataset.csv
df = pd.read_csv('/content/online_gaming_behavior_dataset.csv')
display(df)
# --------------------------------------------------------------

# Understanding The Dataset and Pre-processing
Size_Data = df.size  # The size of the dataset
print("\n Total Number Of Elements : ",Size_Data)
Descibe = df.describe()
print("\n The Description of the dataset :\n",Descibe)

print("The features of our dataset:")
Columns = df.columns
for i in Columns:
  print(i)
# --------------------------------------------------------------

# Data Cleaning
x = df.isnull().sum()
# No NaN values are presemt in dataset
display(x)
df.drop_duplicates(inplace=True)
# No duplicate data in set
# No need for data cleaning.
# --------------------------------------------------------------

# Undertsnading Gamers
# Counting values
print("\n","Total numbers of player in dataset : ",df.shape[0])
GenderCount = df['Gender'].value_counts()
display(GenderCount)
Male = int(GenderCount[0])
Females = int(GenderCount[1])
percentage = (GenderCount / len(df)) * 100
display(percentage.round(2))
# --------------------------------------------------------------

# Gender Distribution
sns.countplot(x='Gender', data=df, palette=['skyblue', 'pink'],edgecolor='black',linewidth=1.5)
plt.xlabel("Gender")
plt.ylabel("Count")
plt.title("Gender Distribution of Gamers",fontsize=16,weight='bold')
plt.gca().set_axisbelow(True)
plt.grid(axis='y', linestyle='-', alpha=1.0)
plt.show()
print("\n")
# --------------------------------------------------------------

# Gender Pie Chart
pie_color = ["skyblue","pink"]
plt.figure(figsize=(7,7))
explode = (0.1,0)
plt.pie(GenderCount,labels=GenderCount.index,autopct='%1.2f%%',explode=explode,startangle=90,colors=pie_color,textprops={'fontsize':12,'weight':'bold'},wedgeprops={'linewidth': 2, 'edgecolor': 'black'})
plt.legend()
plt.title("Gender Distribution of Gamers (%)",fontsize=16,fontweight='bold')
plt.show()
print("\n")
# Male player are more
# aroung 60:40 gender ratio
# --------------------------------------------------------------

# Age Analysis
sns.histplot(df['Age'], kde=True,color='green')
plt.title("Analysis of Age Distribution of Gamers")
plt.gca().set_axisbelow(True)
plt.grid(axis='y', linestyle='-', alpha=1.0)
plt.show()
print("\n")
# --------------------------------------------------------------

# Engagement Level Pie Chart
EngagementLevel = df['EngagementLevel'].value_counts()
percentage_engagement = (EngagementLevel / len(df)) * 100
display(percentage_engagement.round(2))

pie_color = ["yellow",'lightgreen',"red"]
plt.figure(figsize=(7,7))
explode = (0,0.1,0)
plt.pie(EngagementLevel,labels=EngagementLevel.index,autopct='%1.2f%%',startangle=90,colors=pie_color,explode=explode,textprops={'fontsize':12,'weight':'bold'},wedgeprops={'linewidth': 2, 'edgecolor': 'black'})
plt.legend()
plt.title("Engagement Level Distribution of Gamers (%)",fontsize=16,fontweight='bold')
plt.show()
print("\n")
# --------------------------------------------------------------

sns.countplot(x='Gender', hue='EngagementLevel', data=df, palette=["yellow",'lightgreen',"red"],edgecolor='black')
plt.title("Engagement Level Distribution by Gender",fontsize=16,fontweight='bold')
plt.gca().set_axisbelow(True)
plt.grid(axis='y', linestyle='-', alpha=1.0)
plt.show()
print("\n")
# --------------------------------------------------------------

GameType = df['GameGenre'].value_counts()
percentage_game = (GameType / len(df)) * 100
display(percentage_game.round(2))
# As GTA 6 is a action game according to our data analyis Action is still one of the most popular game genre

plt.figure(figsize=(7,7))
explode = (0.15,0,0,0,0)
plt.pie(GameType,labels=GameType.index,autopct='%1.2f%%',startangle=90,colors=['khaki','red','skyblue','lightgreen',"yellow"],explode=explode,textprops={'fontsize':12,'weight':'bold'},wedgeprops={'linewidth': 2, 'edgecolor': 'black'})
plt.legend()
plt.title("Games Genres Distribution (%)",fontsize=16,fontweight='bold')
plt.show()
print("\n")
# --------------------------------------------------------------

Location = df['Location'].value_counts()
percentege_location = (Location / len(df)) * 100
display(percentege_location.round(2))
# Maximum gamers are from USA

plt.figure(figsize=(7,7))
explode = (0.05,0,0,0)
plt.pie(Location,labels=Location.index,autopct='%1.2f%%',startangle=90,colors=['khaki','lightpink','skyblue','lightgreen',"yellow"],explode=explode,textprops={'fontsize':12,'weight':'bold'},wedgeprops={'linewidth': 2, 'edgecolor': 'black'})
plt.legend()
plt.title("Gamers Location wise Distribution (%)",fontsize=16,fontweight='bold')
plt.show()
print("\n")
# --------------------------------------------------------------

# Location And Genre Of Games
sns.countplot(x='Location', hue='GameGenre', data=df, palette=['khaki','red','skyblue','lightgreen',"yellow"],edgecolor='black')
plt.title("Gaming Genre Distribution In Regions",fontsize=16,fontweight='bold')
plt.gca().set_axisbelow(True)
plt.grid(axis='y', linestyle='-', alpha=1.0)
plt.show()
print("\n")
# --------------------------------------------------------------

# Location And Gender
sns.countplot(x='Location', hue='Gender', data=df, palette=["skyblue","pink"],edgecolor='black')
plt.title("Locataion Wise Gender Distribution",fontsize=16,fontweight='bold')
plt.gca().set_axisbelow(True)
plt.grid(axis='y', linestyle='-', alpha=1.0)
plt.show()
print("\n")
# --------------------------------------------------------------

# Gender And Genre Of Games
sns.countplot(x='Gender', hue='GameGenre', data=df, palette=['khaki','red','skyblue','lightgreen',"yellow"],edgecolor='black')
plt.title("Gender Wise Genre Distribution",fontsize=16,fontweight='bold')
plt.gca().set_axisbelow(True)
plt.grid(axis='y', linestyle='-', alpha=1.0)
plt.show()
print("\n")
# --------------------------------------------------------------

# Gaming Time Analysis
# sns.set_style("whitegrid")
sns.histplot(df['PlayTimeHours'],kde=True,color='red',bins=30)
plt.title("Distribution of Gaming Time (Hours)",fontsize=14,fontweight='bold')
plt.xlabel("Play Time (Hours)")
plt.ylabel("Frequency")
plt.gca().set_axisbelow(True)
plt.grid(axis='y', linestyle='-', alpha=1.0)
plt.show()
print("\n")
# --------------------------------------------------------------

# Sessions Per Week Analysis
sns.histplot(df['SessionsPerWeek'], kde=True, color='blue')
plt.title("Gaming Sessions Per Week Distribution",fontsize=16,fontweight='bold')
plt.xlabel("Sessions Per Week")
plt.ylabel("Frequency")
plt.gca().set_axisbelow(True)
plt.grid(axis='y', linestyle='-', alpha=1.0)
plt.show()
print("\n")
# --------------------------------------------------------------

# Session Vs Engagement Analysis
sns.boxplot(x='EngagementLevel',y='SessionsPerWeek',data=df,order=['Low','Medium','High'],palette='coolwarm',linewidth=1.5)
plt.title("Sessions Per Week vs Engagement Level",fontsize=14,fontweight='bold')
plt.xlabel("Engagement Level")
plt.ylabel("Sessions Per Week")
plt.gca().set_axisbelow(True)
plt.grid(axis='y', linestyle='-', alpha=1.0)
plt.show()
print("\n")
# --------------------------------------------------------------

# In Game Purchase
country_total = df.groupby('Location')['InGamePurchases'].sum().sort_values(ascending=True)
plt.figure(figsize=(7,7))
sns.barplot(x=country_total.index,y=country_total.values,palette='coolwarm',edgecolor='black')
plt.title("Total In-Game Purchases by Country",fontsize=14,fontweight='bold')
plt.xlabel("Country")
plt.ylabel("Total Purchases")
plt.xticks(rotation=45)
plt.gca().set_axisbelow(True)
plt.grid(axis='y', linestyle='-', alpha=1.0)
plt.show()
print("\n")
# The probability of highest earning is from USA
# --------------------------------------------------------------

# Engaegement level VS Purchase
sns.barplot(x='EngagementLevel',y='InGamePurchases',data=df,order=['Low','Medium','High'],palette='coolwarm',edgecolor='black')
plt.title("Average In-Game Purchases by Engagement Level",fontsize=14,fontweight='bold')
plt.xlabel("Engagement Level")
plt.ylabel("Average Purchases")
plt.gca().set_axisbelow(True)
plt.grid(axis='y', linestyle='-', alpha=1.0)
plt.show()
print("\n")
# --------------------------------------------------------------

# PlayTimer (Hours) VS Purchase
df['PlayTime_Group'] = pd.cut(df['PlayTimeHours'],bins=[0,5,10,15,20,25],labels=['0-5','5-10','10-15','15-20','20-25'])
avg_purchase = df.groupby('PlayTime_Group')['InGamePurchases'].mean()
sns.lineplot(x=avg_purchase.index,y=avg_purchase.values,marker='s',color='blue')
plt.title("Trend of Purchases vs Play Time",fontsize=16,fontweight='bold')
plt.xlabel("Play Time Range")
plt.ylabel("Average Purchases")
plt.grid(axis='y', linestyle='-', alpha=1.0)
plt.show()
print("\n")
# --------------------------------------------------------------

# Histogram Of Avegeration Playing Duration In Minutes
sns.histplot(df['AvgSessionDurationMinutes'], kde=True, color='skyblue')
plt.title("Session Duration Distribution",fontsize=16,fontweight='bold')
plt.gca().set_axisbelow(True)
plt.grid(axis='y', linestyle='-', alpha=1.0)
plt.show()
print("\n")
# --------------------------------------------------------------

# Game Genre VS Duration
sns.barplot(x='GameGenre',y='AvgSessionDurationMinutes',data=df,palette='coolwarm',edgecolor='black')
plt.title("Average Session Duration by Game Genre",fontsize=16,fontweight='bold')
plt.gca().set_axisbelow(True)
plt.grid(axis='y', linestyle='-', alpha=1.0)
plt.show()
print("\n")
# --------------------------------------------------------------

# Game Genre Playing Time VS Loaction
sns.barplot(x='Location',y='AvgSessionDurationMinutes',hue='GameGenre',data=df,palette='coolwarm',edgecolor='black')
plt.title("Session Duration of Game by Genre Across Locations",fontsize=16,fontweight='bold')
plt.gca().set_axisbelow(True)
plt.grid(axis='y', linestyle='-', alpha=1.0)
plt.show()
print("\n")
# --------------------------------------------------------------

df['Age_Group'] = pd.cut(df['Age'],bins=[15,20,25,30,35,40,45,50],labels=['15-20','20-25','25-30','30-35','35-40','40-45','45-50'])
sns.barplot(x='Age_Group',y='InGamePurchases',hue="GameGenre",data=df,palette='coolwarm',edgecolor='black')
plt.xlabel("Age Group")
plt.ylabel("In Game Purchases By Gamers")
plt.title("Purchases by Age Group",fontsize=16,fontweight='bold')
plt.gca().set_axisbelow(True)
plt.grid(axis='y', linestyle='-', alpha=1.0)
plt.show()
print("\n")
# --------------------------------------------------------------

sns.barplot(x='Age_Group',y='InGamePurchases',hue="Location",data=df,palette='coolwarm',edgecolor='black')
plt.xlabel("Age Group")
plt.ylabel("In Game Purchases By Gamers")
plt.title("Purchases by Age Group",fontsize=16,fontweight='bold')
plt.gca().set_axisbelow(True)
plt.grid(axis='y', linestyle='-', alpha=1.0)
plt.show()
print("\n")
sns.reset_orig()
# --------------------------------------------------------------
```
